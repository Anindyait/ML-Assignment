# Univariate Analysis


```python
accident_severity_counts = df.groupby('Accident_Severity')['Accident_Severity'].count()

sns.countplot(x='Accident_Severity', data=df)
plt.xlabel('Accident Severity')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accident Severity')
plt.show()
```

```python
sns.countplot(x='Did_Police_Officer_Attend_Scene_of_Accident', data=df)
plt.xlabel('Police Officer Attendance')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Police Officer Attendance at Accident Scene')
plt.show()
```

# Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Speed_limit', data=df)
plt.xlabel('Speed Limit')
plt.ylabel('Number of Accidents')
plt.title('Distribution of Accidents by Speed Limit')
plt.show()
```

```python
police_attendance_by_area = df.groupby(['Urban_or_Rural_Area', 'Did_Police_Officer_Attend_Scene_of_Accident'])['Accident_Index'].count().unstack()

police_attendance_by_area.plot(kind='bar', figsize=(10, 6))
plt.title('Police Attendance at Accident Scenes by Urban/Rural Area')
plt.xlabel('Urban or Rural Area')
plt.ylabel('Number of Accidents')
plt.legend(title='Police Officer Attendance')
plt.show()
```

```python
road_casualties = df.groupby('Road_Surface_Conditions')['Number_of_Casualties'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=road_casualties.index, y=road_casualties.values)
plt.xlabel('Road Surface Conditions')
plt.ylabel('Average Number of Casualties')
plt.title('Road Surface Conditions vs. Number of Casualties')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
casualties_by_road_type = df.groupby('Road_Type')['Number_of_Casualties'].sum()
plt.figure(figsize=(20, 4))
plt.bar(casualties_by_road_type.index, casualties_by_road_type.values)
plt.xlabel('Road Type')
plt.ylabel('Number of Casualties')
plt.title('Casualties by Road Type')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Year'] = df['Date'].dt.year

casualties_per_year = df.groupby('Year')['Number_of_Casualties'].sum()

sns.barplot(x=casualties_per_year.index, y=casualties_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Casualties')
plt.title('Casualties per Year')
plt.show()
```

```python
casualties_per_speed_limit = df.groupby('Speed_limit')['Number_of_Casualties'].sum()

sns.barplot(x=casualties_per_speed_limit.index, y=casualties_per_speed_limit.values)
plt.xlabel('Speed Limit')
plt.ylabel('Number of Casualties')
plt.title('Number of Casualties vs Speed Limit')
plt.show()
```
